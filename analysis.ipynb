```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../test/data.csv") 
df.floorType = df.floorType.replace('顶','4')
df.floorType = df.floorType.replace('高','3')
df.floorType = df.floorType.replace('中','2')
df.floorType = df.floorType.replace('低','1')
df.floorType = df.floorType.replace('底','0')
df.floorType = df.floorType.astype('int64')
```

```python
df.head()
```

```python
df_2016 = df.loc[df.tradeyear == 2016]
sns.distplot(a = df_2016['square'], kde = False)  #plot the effect of square on the amount of housing traded
plt.xlabel('Square') 
plt.ylabel('Amount_of_Housing_Traded')
```

```python
sns.scatterplot(x = df_2016['square'], y = df_2016['followers'])  #plot the effect of square on followers
plt.xlabel('Square')
plt.ylabel('Followers')
```

```python
sns.distplot(a = df_2016['livingRoom'], kde = False, label = 'LivingRoom_Number')  #plot the effect of room number on the amount of housing traded
sns.distplot(a = df_2016['bathRoom'], kde = False, label = 'BathRoom_Number')
plt.xlabel('Room_Number')
plt.ylabel('Amount_of_Housing_Traded')
plt.legend()
```

```python
livingRoomNum = df_2016.groupby('livingRoom').price.mean()   #plot the effect of living room number on the mean price 
sns.barplot(x = livingRoomNum.index, y = livingRoomNum)
plt.xlabel('LivingRoom_Number')
plt.ylabel('Mean_Price')
```

```python
sns.distplot(a = df_2016['floorType'], kde = False)  #plot the effect of floor type on the amount of housing traded
plt.xlabel('Floor_Type')
plt.ylabel('Amount_of_Housing_Traded')
```
